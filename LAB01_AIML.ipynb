```python
people =['Ann','Brandon','Chen','Daemon']
age=[21,12,32,45]
weight=[55,45,65,14]
height=[160,158,149,156]
import matplotlib.pyplot as plt
plt.scatter(age, height)
plt.show()
```

```python
plt.figure(figsize=(12,4))
plt.title("People's weight in kgs",fontsize =16,fontstyle='italic')
#Main plot function 'bar'
plt.bar(x=people,height=weight,width=0.6,color='orange',edgecolor='k',alpha=0.6)
plt.xlabel("People",fontsize=15)
plt.xticks(fontsize=14,rotation=0)
plt.yticks(fontsize=14)
plt.ylabel("Weight (in kgs)",fontsize=15)
plt.show()
```

```python
import numpy as np
plt.figure(figsize=(4,5))
plt.title("People's weight in kgs",fontsize =16)
plt.hist(weight,color='orange',edgecolor='k',alpha=0.75,bins=5)
plt.xlabel("People",fontsize=15)
plt.xticks(fontsize=14)
plt.yticks(fontsize=15)

plt.show()
```
